```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.data.shape
```

```python
iris_data_jitter = iris.data + 0.1 * np.random.rand(150, 4) - 0.05
```

```python
iris_df = pd.DataFrame(iris_data_jitter, columns=iris.feature_names)
iris_df['class'] = [iris.target_names[class_id] for class_id in iris.target]
```

```python
iris_df.head()
```

```python
sns.pairplot(data=iris_df, hue='class')
```

```python
iris_train, iris_test = train_test_split(iris_df, test_size=0.5, random_state=42)
```

```python
col_x = 'petal width (cm)'
col_y = 'sepal width (cm)'

for cl in iris.target_names:
    X = iris_train[iris_train['class'] == cl]
    plt.plot(X[col_x], X[col_y], 'o', label=cl)
plt.xlim([0, 3])
plt.ylim([2, 4.5])
plt.legend(loc='upper right')
plt.xlabel(col_x)
plt.ylabel(col_y)
plt.title("Iris dataset: training")
plt.savefig("iris_train.pdf")
```

```python
for cl in iris.target_names:
    X = iris_test[iris_test['class'] == cl]
    plt.plot(X[col_x], X[col_y], 'o', label=cl)
plt.xlim([0, 3])
plt.ylim([2, 4.5])
plt.legend(loc='upper right')
plt.xlabel(col_x)
plt.ylabel(col_y)
plt.title("Iris dataset: validation")
plt.savefig("iris_test.pdf")
```
